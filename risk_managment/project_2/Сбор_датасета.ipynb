```python
brent = pd.read_csv('data/brent.csv')[['Дата', 'Цена']].rename(columns={'Цена':'Brent'})
usd = pd.read_csv('data/usd.csv')[['Дата', 'Цена']].rename(columns={'Цена':'USD'})
eur = pd.read_csv('data/eur.csv')[['Дата', 'Цена']].rename(columns={'Цена':'EUR'})
rts = pd.read_csv('data/rts.csv')[['Дата', 'Цена']].rename(columns={'Цена':'RTS'})
moex = pd.read_csv('data/moex.csv')[['Дата', 'Цена']].rename(columns={'Цена':'MOEX'})
moex.shape, brent.shape, eur.shape, usd.shape, rts.shape
```

```python
all_data = pd.merge(brent, usd, on='Дата', how='outer')
all_data = pd.merge(all_data, eur, on='Дата', how='outer')
all_data = pd.merge(all_data, rts, on='Дата', how='outer')
all_data = pd.merge(all_data, moex, on='Дата', how='outer')
all_data['Дата'] = pd.to_datetime(all_data['Дата'], format = '%d.%m.%Y')
all_data = all_data.rename(columns={'Дата':'date'})
```

# Дополнительные риск-факторы

```python

risk_factors = {'GAS': 'NG=F',
                'GOLD': 'MGC=F',
                'SILVER': 'SI=F',
    }
df_risk_factors = {}
for key, value in risk_factors.items():
    df_risk_factors[key] = yf.download(
      value,
      start='2020-01-01', 
      end='2024-01-01'
      )['Adj Close']
df_risk_factors = pd.DataFrame(df_risk_factors)
df_risk_factors.reset_index(inplace=True)
df_risk_factors.rename(columns={'Date':'date'}, inplace=True)
df_risk_factors
```

```python
final_data = pd.merge(all_data, stocks, on='date', how='outer')
final_data = pd.merge(final_data, ofz, on='date', how='outer')
final_data = pd.merge(final_data, int_rates, on='date', how='outer')
final_data = pd.merge(final_data, df_risk_factors, on='date', how='outer')
final_data
```

```python
for col in final_data.columns[1:]:
    if final_data[col].dtype != 'float64':    
        final_data[col] = final_data[col].str.replace('.', '')
        final_data[col] = final_data[col].str.replace(',', '.').astype(float)
final_data.dtypes
```

```python
final_data = final_data.sort_values('date')
```

```python
final_data.fillna(method='ffill', inplace=True)
final_data.fillna(method='bfill', inplace=True)
```

```python
final_data['Brent'] *= final_data['USD']
final_data['RTS'] *= final_data['USD']
final_data['GAS'] *= final_data['USD']
final_data['GOLD'] *= final_data['USD']
final_data['SILVER'] *= final_data['USD']
final_data
```

```python
final_data.sort_values(by=['date'], inplace=True)
final_data.set_index('date', inplace=True)
```

```python
final_data.to_csv('data/risks_dataset.csv')
```
