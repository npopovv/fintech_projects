```python
def fill_missing_values_optimized(df, unique_threshold=18, untouched_cols=['issue_d', 'last_credit_pay_date']):
    """
    Fill missing values with the mean for high unique columns that contain NaN values
    and return the entire DataFrame, keeping the specified columns untouched.

    Args:
        df (pandas.DataFrame): The input DataFrame.
        unique_threshold (int): The threshold for the number of unique values in a column.
        untouched_cols (list): A list of column names to be kept untouched.

    Returns:
        pandas.DataFrame: The DataFrame with filled missing values for the specified columns.
    """
    low_unique_cols = df.nunique() < unique_threshold
    low_unique_cols = low_unique_cols[low_unique_cols].index.tolist()

    #у бывших категориальных признаков не оказалось пропусков, поэтому замена проп. значений
    #была нужна только для тру-численных фичей

    high_unique_cols = df.columns.difference(low_unique_cols)
    high_unique_cols = high_unique_cols.difference(untouched_cols)

    high_unique_nan_cols = df[high_unique_cols].isnull().any()
    high_unique_nan_cols = high_unique_nan_cols[high_unique_nan_cols].index.tolist()

    column_means = df[high_unique_nan_cols].mean()
    df_filled = df[high_unique_nan_cols].fillna(column_means)

    df[high_unique_nan_cols] = df_filled

    return df
```

```python
DT = DecisionTreeClassifier(criterion='gini', random_state=100, max_depth=5,
                            class_weight='balanced', min_samples_leaf=5)
RF = RandomForestClassifier(criterion='gini', random_state=100, max_depth=5,
                            class_weight='balanced', min_samples_leaf=5, n_estimators=20)
Bagged = BaggingClassifier(n_estimators=50)
AdaBoost = AdaBoostClassifier(n_estimators=50)
LR = LogisticRegression(solver='lbfgs', max_iter=1000)
XGB = xgb.XGBClassifier(n_estimators=50)
```

```python
models = []

models.append(('DT', DT))
models.append(('RandomForest', RF))
models.append(('Bagged', Bagged))
models.append(('AdaBoost', AdaBoost))
models.append(('LogisticRegression', LR))
models.append(('XGBoost', XGB))
```

```python
# evaluating each model in turn using KFold CV with 3 splits
results = []
names = []
for name, model in models:
    kfold = model_selection.KFold(shuffle=True, n_splits=3, random_state=0)
    cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring='f1_micro')
    results.append(cv_results)
    names.append(name)
    print("%s: %f (%f)" % (name, 1 - np.mean(cv_results), np.std(cv_results, ddof=1)))
```

```python
fig = plt.figure()
sns.set_style('whitegrid')
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
sns.set(rc={'figure.figsize': (50, 50)})
plt.xticks(rotation='horizontal')
plt.ylabel('Weighted F1 Score')
ax.set_xticklabels(names)
plt.show()
```
